```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# 데이터 크롤링

```python
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# search = input("Enter search term: ")
search = "당근마켓"
url = "https://search.naver.com/search.naver?"

params = {
    "where": "news",
    "query": search,
    "start": 0,
    "nso": "so:r,p:7d,a:all",  # 최근 7일
    # "sm": "tab_opt",
    # "sort": "0",
    # "photo": "0",
    # "field": "0",
    # "pd": "0",
    # "ds": "",
    # "de": "",
    # "cluster_rank": "29",
    # "mynews": "0",
    # "office_type": "0",
    # "office_section_code": "0",
    # "news_office_checked": "",
    # "nso": "",
}

re_url = requests.get(url=url, headers=header, params=params)
html = BeautifulSoup(re_url.text, "html.parser")

html
```

```python
# html body
articles = html.select("ul.type01 > li")

articles

# 전체 기사 수
# totalCount = html.select("div.section_head > div.title_desc > span")[0].text.split(' / ')[1][:-1]
```

```python
articles = html.select('a[data-heatmap-target=".nav"]')
articles
```

```python
article = articles[0]
title = article.text

title, len(articles)
```

```python
data = pd.DataFrame(columns=["title", "link"])

for article in articles:
    new_data = [article.text, article["href"]]
    data.loc[len(data)] = new_data

data
```

```python
data.to_csv("../_data/news_crawling.csv", index=False)
```
